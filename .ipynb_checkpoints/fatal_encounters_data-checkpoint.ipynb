```python
fe = pd.read_csv("fatal_encounters_form_responses.csv")
```

```python
fe.shape
```

```python
fe = fe.iloc[:,:23]
```

```python
fe.columns
```

```python
fe.rename(columns={'Agency responsible for death': 'agency', "Date of injury resulting in death (month/day/year)": "date", "Cause of death":"cause"}, inplace=True)
```

```python
fe['time'] = pd.to_datetime(fe['date'])
```

```python
fe["year"] = fe["time"].dt.year
```

```python
year_counts = fe["year"].value_counts()
```

```python
year_counts.reindex(year_counts.index)
```

```python
plt.figure(figsize=(36, 12))
plt.bar(year_counts.index[1:], year_counts[1:], width=0.55, align="center")
plt.xlabel("Year", size=30)
plt.ylabel("Number of deaths in a year", size=30)
plt.xticks(range(2000,2017), size=20,rotation="vertical")
# plt.tight_layout()
plt.title("Number of police-related fatalities by year", size=40)
plt.show()
```
